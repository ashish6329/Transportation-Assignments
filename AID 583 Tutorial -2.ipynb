```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
print("Name : Ashish Varshney")
print("Enrollment No: 21565004")
```

```python

print("Ques a:Prepare a heatmap to show which variables have the highest and second highest numberof missing data.\n\n")
ak=pd.read_csv("train.csv")
ak.isna().sum()
#ak.loc[:, ak.isnull().any()].columns
sns.heatmap(ak.isnull(),yticklabels = False, cbar = False, cmap = 'summer')
plt.title('heatmap to show highest and second heighest missing data',size=11)
ak.loc[:, ak.isnull().any()].columns
plt.xticks(rotation=45)
plt.show()
```

```python
print("Ques b:Prepare a countplot showing how many people survived and how many did not.\n\n")
ak=pd.read_csv("train.csv")
sur=ak.Survived
sns.set_theme(style="darkgrid")
sns.countplot(x="Survived", data=ak,palette="Set1")
plt.title("Count plot Survived Vs Unsurvived",size=14)
plt.xticks(rotation=0)
plt.yticks(rotation=0)
plt.show()
print(ak.Survived.value_counts())
```

```python
print("Ques c: Prepare another countplot showing how many male and female survived vs. not survived\n\n")
ak=pd.read_csv("train.csv")
sur=ak.Survived
sex=ak.Sex
sns.set_theme(style="darkgrid")
#sns.countplot(x="Survived", data=ak)
sns.countplot(x='Sex',hue="Survived",data=ak,palette="Set2")
plt.title("Count plot of Male and Female Survived Vs Unsurvived",size=14)
plt.xticks(rotation=0)
plt.yticks(rotation=0)
plt.show()
print(ak.Survived.value_counts())
print(ak.Sex.value_counts())

print(ak.Survived.groupby(ak.Sex).value_counts())
```

```python

print("Ques d: For the countplot in c, change the colour of the bars to red for male and blue for female\n\n")
ak=pd.read_csv("train.csv")
sur=ak.Survived
sex=ak.Sex
sns.set_theme(style="darkgrid")
#sns.countplot(x="Survived", data=ak)
sns.countplot(x='Survived',hue="Sex",data=ak,palette=['Red','Blue'])
plt.title("Count plot of Male and Female Survived Vs Unsurvived",size=14)
plt.show()
print(ak.Survived.value_counts())
print(ak.Sex.value_counts())
print(ak.Survived.groupby(ak.Sex).value_counts())
```

```python

print('Ques e:Prepare a countplot showing how many survived vs. not survived in reference topassenger class?\n\n')
ak=pd.read_csv("train.csv")
sur=ak.Survived
sex=ak.Sex
Passenger_class=ak.Pclass
sns.set_theme(style="darkgrid")
#sns.countplot(x="Survived", data=ak)
sns.countplot(x='Pclass',hue="Survived",data=ak)
plt.title("Count plot of Passenger class Survived Vs Unsurvived ",size=14)
plt.xticks(rotation=0)
plt.yticks(rotation=0)
plt.show()
print(ak.Survived.value_counts())
print(ak.Pclass.value_counts())
print(ak.Survived.groupby(ak.Pclass).value_counts())
```

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
print("Ques f:Show the distribution of the age of the people on the Titanic with 30 beans.\n\n")
ak=pd.read_csv("train.csv")
data=ak.Age
sns.distplot(data,bins=30)
plt.title('Average age of Passenger',size=11)
plt.show()
print("Average age:",data.mean())
```

```python

print("Ques g:Prepare a countplot to show the number of sibling/spouse of passengers on-board.\n\n")
ak=pd.read_csv("train.csv")
sur=ak.Survived
sns.set_theme(style="darkgrid")
sns.countplot(x="SibSp", data=ak,palette="Set2")
plt.title("Count plot Survived Vs Unsurvived",size=14)
plt.xticks(rotation=0)
plt.yticks(rotation=0)
plt.xlabel('Numer of Siblings/Spouses Aboard')
plt.show()
print(ak.SibSp.value_counts())
```
